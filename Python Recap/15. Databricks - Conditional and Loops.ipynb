```python
#fizzbuzz lab

for i in range(1,101):
    if i % 3 == 0 and i % 5 == 0:
        print("Fizzbuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
#fizzbuzz lab

for num in range(1,101):
    output = ""
    
    if num % 3 == 0:
        output += "Fizz"
    if num % 5 == 0:
        output += "Buzz"
    if num % 3 != 0 and num % 5 != 0:
        output = f"{num}"
    print(output)
```
